```python
import os
```

```python
os.environ["KERAS_BACKEND"] = "plaidml.keras.backend"
```

```python
from imagecluster import calc as ic
from imagecluster import postproc as pp
```

```python
DATA_DIR = "datasets/downloads"
```

```python
for subdir in sorted(os.listdir(DATA_DIR)):
    bird_path = os.path.join(DATA_DIR, subdir)
    if os.path.isdir(bird_path):
        bird_name = os.path.basename(subdir)
        print("[INFO] :: Clustering {0}".format(bird_name))
        ias = ic.image_arrays(bird_path, size=(224,224))
        model = ic.get_model()
        fps = ic.fingerprints(ias, model)
        fps = ic.pca(fps, n_components=0.95)
        clusters = ic.cluster(fps, sim=0.5)
        clusters_path = os.path.join("birds_clusters",bird_name)
        print("[INFO] :: Saving clusters at {0}".format(clusters_path))
        pp.make_links(clusters, clusters_path)
```

```python
# Create image database in memory. This helps to feed images to the NN model
# quickly.
BIRD_NAME = "Hohltaube"

ias = ic.image_arrays('datasets/downloads/{0}'.format(BIRD_NAME), size=(224,224))
```

```python
# Create Keras NN model.
model = ic.get_model()
```

```python
# Feed images through the model and extract fingerprints (feature vectors).
fps = ic.fingerprints(ias, model)
```

```python
# Optionally run a PCA on the fingerprints to compress the dimensions. Use a
# cumulative explained variance ratio of 0.95.
fps = ic.pca(fps, n_components=0.95)
```

```python
# Run clustering on the fingerprints.  Select clusters with similarity index
# sim=0.5
clusters = ic.cluster(fps, sim=0.5)
```

```python
# Create dirs with links to images. Dirs represent the clusters the images
# belong to.
pp.make_links(clusters, 'pics/{0}/clusters'.format(BIRD_NAME))
```

```python
# Plot images arranged in clusters.
pp.visualize(clusters, ias)
```
